```python
##polynomial regression &  linear regression
import numpy as np
import matplotlib.pyplot as plt
year = [2,4,6,8,11,12,14,15,18,19]
marks = [100,50,60,60,70,80,50,40,90,100]
mymodel = np.poly1d(np.polyfit(year, marks, 4))
myline = np.linspace(2, 19, 100)
plt.scatter(year, marks)
plt.plot(myline, mymodel(myline))
plt.xlabel('Year')
plt.ylabel('Marks')
plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy import stats
year=[2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]
marks=[85,89,91,95,82,83,87,94,86,98]
slope,intercept,r,p,std_err= stats.linregress(year,marks)
def my_func(year):
    return slope*year+intercept
mymodel=list(map(my_func,year))
plt.scatter(year,marks)
plt.plot(year,mymodel)
plt.show()
```
