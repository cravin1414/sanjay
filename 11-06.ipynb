```python
##normalization in feature scaling using python ai&ml

import pandas as pd
```

```python
employess={"emp_name":["cravin","rahul","dharani","sathish"],"Age":[12,31,21,32],"salary":[20000,13000,52000,30000],"quit_job":["yes","no","no","yes"]}
df=pd.DataFrame.from_dict(employess)
```

```python
df
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
scaler.fit(df[["Age","salary"]])
```

```python
scaler_values=scaler.transform(df[["Age","salary"]])
```

```python
scaler_values
```
