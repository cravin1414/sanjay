```python
### exception handling
try:
    a=int(input("enter the num:"))
    b=int(input("enter the num:"))
    print(a/b)
except ZeroDivisionError as e:
    print("ZeroDivisionError",e)
finally:
    print("done")
    
```

```python
###multiple exception block
try:
    a=int(input("enter the num:"))
    b=int(input("enter the number:"))
    c=int(input("enter the number:"))
    print(a+b/c)
except Exception as e:
    print("Exception",e)
except ValueError as e:
    print("ValueError",e)
except ZeroDivisionError as e:
    print("ZeroDivisionError",e)
finally:
    print("iam done")
    
    
```

```python
##raise keyword
def person(age):
    if age>18:
        print("major")
    else:
        raise ZeroDivisionError("minor")

try:
    person(35)
    person(40)
    person(12)
    person(67)
except Exception as e:
    print(e)
    
```
